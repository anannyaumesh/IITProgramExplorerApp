```python
import tkinter as tk
from tkinter import ttk
import folium
from tkinterweb import HtmlFrame

class ProgramExplorerApp:
    def __init__(self, master):
        self.master = master
        self.master.title("IIT-NIT Program Explorer")
        self.master.geometry("800x600")
       
        # Label for the title
        self.title_label = tk.Label(self.master, text="IIT-NIT Program Explorer", font=("Helvetica", 20))
        self.title_label.pack(pady=20)
       
        # Paragraph explaining the purpose
        self.purpose_text = "A tool to help aspirants of different IIT/NITs understand their cut off based on category."
        self.purpose_label = tk.Label(self.master, text=self.purpose_text, wraplength=600, justify="center")
        self.purpose_label.pack(pady=10)
       
        # Buttons for FAQ and MAP
        self.faq_button = ttk.Button(self.master, text="FAQ", command=self.open_faq_window)
        self.faq_button.pack(side=tk.LEFT, padx=20)
       
        self.map_button = ttk.Button(self.master, text="MAP", command=self.open_map_window)
        self.map_button.pack(side=tk.RIGHT, padx=20)
   
    def open_faq_window(self):
        # Function to open FAQ window
        faq_window = tk.Toplevel(self.master)
        faq_window.title("FAQ")
        faq_window.geometry("400x300")
       
        faqs = [
            "What is the application about?",
            "How can I use the MAP feature?",
            "Can I customize the MAP view?"
        ]
       
        self.selected_faq = tk.StringVar()
       
        for idx, faq in enumerate(faqs):
            tk.Radiobutton(faq_window, text=faq, variable=self.selected_faq, value=str(idx)).pack(anchor='w')
       
        submit_button = tk.Button(faq_window, text="Submit", command=self.submit_faq)
        submit_button.pack(pady=10)
   
    def submit_faq(self):
        selected_index = int(self.selected_faq.get())
        faq_answers = [
            "A1: This application helps aspirants of different IIT/NITs understand their cut off based on category.",
            "A2: Click on the MAP button to open an interactive map showing the location of IIT/NITs.",
            "A3: Yes, you can zoom in/out and pan the map to explore different areas."
        ]
       
        answer_window = tk.Toplevel(self.master)
        answer_window.title("FAQ Answer")
        answer_window.geometry("400x200")
       
        answer_label = tk.Label(answer_window, text=faq_answers[selected_index], font=("Helvetica", 12), wraplength=380)
        answer_label.pack(pady=20)
   
    def open_map_window(self):
        # Function to open MAP window
        map_window = tk.Toplevel(self.master)
        map_window.title("MAP")
        map_window.geometry("800x600")
       
        # Create a Folium map
        m = folium.Map(location=[20.5937, 78.9629], zoom_start=5)
       
        # Add marker for IIT Bombay
        folium.Marker([19.1334, 72.9133], popup='<b>IIT Bombay</b>').add_to(m)
       
        # Add marker for IIT Delhi
        folium.Marker([28.7041, 77.1025], popup='<b>IIT Delhi</b>').add_to(m)
       
        # Add marker for IIT Madras
        folium.Marker([13.0064, 80.2566], popup='<b>IIT Madras</b>').add_to(m)
       
        # Convert the Folium map to HTML string
        map_html = m._repr_html_()
       
        # Create a WebView to display the map
        map_frame = HtmlFrame(map_window, horizontal_scrollbar="auto", vertical_scrollbar="auto")
        map_frame.set_content(map_html)
        map_frame.pack(fill="both", expand=True)


# Create the main application window
root = tk.Tk()
app = ProgramExplorerApp(root)
root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk
import tkintermapview
import mysql.connector
from PIL import Image, ImageTk


class ProgramExplorerApp:
    def __init__(self, master):
        self.master = master
        self.master.title("IIT-NIT Program Explorer")
        self.master.geometry("800x600")

        # Label for the title
        self.title_label = tk.Label(self.master, text="IIT-NIT Program Explorer", font=("Helvetica", 20))
        self.title_label.pack(pady=20)

        # Paragraph explaining the purpose
        self.purpose_text = "A tool to help aspirants of different IIT/NITs understand their cut off based on category."
        self.purpose_label = tk.Label(self.master, text=self.purpose_text, wraplength=600, justify="center")
        self.purpose_label.pack(pady=10)

        # Slide show variables
        self.image_list = []  # List to store image paths
        self.image_index = 0  # Current image index
        self.image_label = tk.Label(self.master)  # Label to display slideshow image
        self.image_label.pack(pady=20)  # Position the label

        # Load images (replace these with your actual image paths)
        for image_path in ["/Users/anannyaumesh/Desktop/pics/iit_bombay.jpeg", "/Users/anannyaumesh/Desktop/pics/iit_delhi.jpeg", "/Users/anannyaumesh/Desktop/pics/random.png"]:
            self.image_list.append(ImageTk.PhotoImage(Image.open(image_path)))

        # Start the slideshow loop
        self.show_slideshow()

        # Buttons for FAQ, Find Your Rank, and MAP
        self.faq_button = ttk.Button(self.master, text="FAQ", command=self.open_faq_window)
        self.faq_button.pack(side=tk.LEFT, padx=20, pady=10)

        self.find_rank_button = ttk.Button(self.master, text="Find Your Rank", command=self.open_find_rank_window)
        self.find_rank_button.pack(side=tk.LEFT, padx=20, pady=10)

        self.map_button = ttk.Button(self.master, text="MAP", command=self.open_map_window)
        self.map_button.pack(side=tk.RIGHT, padx=20, pady=10)

    def open_map_window(self):
        map_window = tk.Toplevel(self.master)
        map_window.title("IIT/NIT Map")
        map_window.geometry("900x700")

        my_label = tk.LabelFrame(map_window)
        my_label.pack(pady=20)

        map_widget = tkintermapview.TkinterMapView(my_label, width=800, height=600, corner_radius=0)

        # Create a function to open a blank window when a marker is clicked
        def open_blank_window(name):
            blank_window = tk.Toplevel(map_window)
            blank_window.title(f"{name} Information")
            blank_window.geometry("400x300")  # You can adjust this size

        # Add markers with the command to open blank windows
        map_widget.set_position(12.9915, 80.2337, marker=True, text="IIT Madras",
                                command=lambda name="IIT Madras": open_blank_window(name))  # Capture name
        map_widget.set_position(28.5457, 77.1928, marker=True, text="IIT Delhi",
                                command=lambda name="IIT Delhi": open_blank_window(name))  # Capture name
        map_widget.set_position(26.4710, 73.1134, marker=True, text="IIT Jodhpur",
                                command=lambda name="IIT Jodhpur": open_blank_window(name))
        map_widget.set_position(22.3149, 87.3105, marker=True, text="IIT Kharagpur",
                                command=lambda name="IIT Kharagpur": open_blank_window(name))
        map_widget.set_position(19.1334, 72.9133, marker=True, text="IIT Bombay",
                                command=lambda name="IIT Bombay": open_blank_window(name))

        # Set zoom level
        map_widget.set_zoom(5)

        map_widget.pack()

    def open_find_rank_window(self):
        # Open the "Find Your Rank" window
        find_rank_window = tk.Toplevel(self.master)
        find_rank_window.title("Find Your Rank")
        find_rank_window.geometry("400x400")

        # Connect to the database
        self.connection = mysql.connector.connect(
            host="localhost",
            user="root",
            password="$$jkahds76!",
            database="iit_nit_cutoff"
        )
        self.cursor = self.connection.cursor()  # Create a cursor object

        # Get distinct rounds
        round_label = tk.Label(find_rank_window, text="Choose Round:")
        round_combobox = ttk.Combobox(find_rank_window, values=self.get_rounds(), state="readonly")
        round_combobox.bind("<<ComboboxSelected>>", lambda event: self.update_listboxes(round_combobox.get()))

        # Initialize listboxes with empty values
        self.institute_type_listbox = tk.Listbox(find_rank_window)
        self.institute_names_listbox = tk.Listbox(find_rank_window)
        self.program_names_listbox = tk.Listbox(find_rank_window)
        self.categories_listbox = tk.Listbox(find_rank_window)

        explore_button = tk.Button(find_rank_window, text="Explore",
                                   command=lambda: self.explore_round(self.cursor, round_combobox))

        # Place widgets on the window
        round_label.pack(pady=10)
        round_combobox.pack(pady=5, padx=10)

        self.institute_type_listbox.pack(pady=5, padx=10)
        self.institute_names_listbox.pack(pady=5, padx=10)
        self.program_names_listbox.pack(pady=5, padx=10)
        self.categories_listbox.pack(pady=5, padx=10)

        explore_button.pack(pady=10)

    def show_slideshow(self):
        self.image_label.config(image=self.image_list[self.image_index])
        self.image_index = (self.image_index + 1) % len(self.image_list)  # Wrap around
        self.master.after(3000, self.show_slideshow)

    def open_faq_window(self):
        # Function to open FAQ window
        faq_window = tk.Toplevel(self.master)
        faq_window.title("FAQ")
        faq_window.geometry("400x300")

        faqs = [
            "What is the application about?",
            "How can I use the MAP feature?",
            "Can I customize the MAP view?"
        ]

        selected_faq = tk.StringVar()

        for idx, faq in enumerate(faqs):
            tk.Radiobutton(faq_window, text=faq, variable=selected_faq, value=str(idx)).pack(anchor='w')

        submit_button = tk.Button(faq_window, text="Submit", command=self.submit_faq)
        submit_button.pack(pady=10)

    def submit_faq(self):
        selected_index = self.selected_faq.get()
        if selected_index:
            selected_index = int(selected_index)
            faq_answers = [
                "A1: This application helps aspirants of different IIT/NITs understand their cut off based on category.",
                "A2: Click on the MAP button to open an interactive map showing the location of IIT/NITs.",
                "A3: Yes, you can zoom in/out and pan the map to explore different areas."
            ]

            answer_window = tk.Toplevel(self.master)
            answer_window.title("FAQ Answer")
            answer_window.geometry("400x200")

            answer_label = tk.Label(answer_window, text=faq_answers[selected_index], font=("Helvetica", 12),
                                    wraplength=380)
            answer_label.pack(pady=20)

    def get_rounds(self):
        """Retrieves distinct round numbers from student_data and returns a list."""
        query = "SELECT DISTINCT round_no FROM student_data ORDER BY round_no ASC;"
        self.cursor.execute(query)
        round_rows = self.cursor.fetchall()
        round_list = [row[0] for row in round_rows]  # Extract distinct round numbers
        return round_list

    def get_institute_types(self):
        """Returns a list of distinct institute types."""
        query = "SELECT DISTINCT institute_type FROM student_data"
        self.cursor.execute(query)
        institute_type_rows = self.cursor.fetchall()
        institute_type_list = [row[0] for row in institute_type_rows]
        return institute_type_list

    def get_institute_names(self, selected_round=None, selected_institute_type=None):
        """Returns a list of distinct institute names based on filters (round & institute type)."""
        query = """SELECT DISTINCT institute_name FROM student_data"""
        filters = []
        if selected_round:
            filters.append(f"round_no = {selected_round}")
        if selected_institute_type:
            filters.append(f"institute_type = '{selected_institute_type}'")
        if filters:
            query += " WHERE " + " AND ".join(filters)
        self.cursor.execute(query)
        institute_name_rows = self.cursor.fetchall()
        institute_name_list = [row[0] for row in institute_name_rows]
        return institute_name_list

    def get_program_names(self, selected_round=None, selected_institute_name=None):
        """Returns a list of program names with details based on filters (round & institute name)."""
        query = """
            SELECT DISTINCT program_name
            FROM student_data
        """
        filters = []
        if selected_round:
            filters.append(f"round_no = {selected_round}")
        if selected_institute_name:
            filters.append(f"institute_name = '{selected_institute_name}'")
        if filters:
            query += " WHERE " + " AND ".join(filters)
        query += " GROUP BY program_name, program_duration, degree_short;"
        self.cursor.execute(query)
        program_name_rows = self.cursor.fetchall()
        program_name_list = [row[0] for row in program_name_rows]
        return program_name_list

    def get_categories(self, selected_round=None, selected_program_name=None):
        """Returns a list of distinct categories based on filters (round & program name)."""
        query = "SELECT DISTINCT category FROM student_data"
        filters = []
        if selected_round:
            filters.append(f"round_no = {selected_round}")
        if selected_program_name:
            filters.append(f"program_name = '{selected_program_name}'")
        if filters:
            query += " WHERE " + " AND ".join(filters)
        self.cursor.execute(query)
        category_rows = self.cursor.fetchall()
        category_list = [row[0] for row in category_rows]
        return category_list

    def update_listboxes(self, selected_round):
        # Get institute types based on the selected round (if any)
        institute_types = self.get_institute_types(selected_round)  # Call with only selected_round

        # Clear the institute type listbox
        self.institute_type_listbox.delete(0, tk.END)

        # Populate the institute type listbox
        for institute_type in institute_types:
            self.institute_type_listbox.insert(tk.END, institute_type)

        # Now populate subsequent listboxes based on selections (if any)
        if selected_round:
            self.update_institute_names(self.cursor, selected_round)
            self.update_program_names(self.cursor, selected_round)
            self.update_categories(self.cursor, selected_round)

    # Update institute names function (assuming similar logic as get_institute_names)
    def update_institute_names(self, cursor, selected_round, selected_institute_type=None):
        institute_names = self.get_institute_names(selected_round, selected_institute_type)
        self.institute_names_listbox.delete(0, tk.END)
        for institute_name in institute_names:
            self.institute_names_listbox.insert(tk.END, institute_name)

    # Update program names function (assuming similar logic as get_program_names)
    def update_program_names(self, cursor, selected_round, selected_institute_type=None):
        program_names = self.get_program_names(selected_round, selected_institute_type)
        self.program_names_listbox.delete(0, tk.END)
        for program_name in program_names:
            self.program_names_listbox.insert(tk.END, program_name)

    # Update categories function (assuming similar logic as get_categories)
    def update_categories(self, cursor, selected_round, selected_institute_type=None):
        categories = self.get_categories(selected_round, selected_institute_type)
        self.categories_listbox.delete(0, tk.END)
        for category in categories:
            self.categories_listbox.insert(tk.END, category)

    def explore_round(self, cursor, combobox):
        selected_round = combobox.get()
        selected_institute_type = self.institute_type_listbox.get(self.institute_type_listbox.curselection())
        selected_institute_name = self.institute_names_listbox.get(self.institute_names_listbox.curselection())
        selected_program_name = self.program_names_listbox.get(self.program_names_listbox.curselection())
        selected_category = self.categories_listbox.get(self.categories_listbox.curselection())


        if selected_round:
            round_detail_window = tk.Toplevel(self.master)  # Opens a new Window
            round_detail_window.title(f"Round Detail: {selected_round}")

            # Display round details in a treeview
            # define columns (Ensure column names match heading text exactly)
            result_column = ("institute_name", "program_name", "quota", "category", "pool", "opening_rank",
                             "closing_rank")
            tree = ttk.Treeview(round_detail_window, columns=result_column, show='headings')
            tree.heading("institute_name", text="Institute")
            tree.heading("program_name", text="Program")
            tree.heading("quota", text="Quota")
            tree.heading("category", text="Category")
            tree.heading("pool", text="Pool")
            tree.heading("opening_rank", text="Opening Rank")
            tree.heading("closing_rank", text="Closing Rank")

            # Improved query with filters (assuming no JOINs needed)
            query = f"""
                SELECT institute_name, program_name, quota, category, pool, 
                       opening_rank, closing_rank
                FROM student_data
                WHERE round_no = {selected_round}
            """
            filters = []
            if selected_institute_type:
                filters.append(f"institute_type = '{selected_institute_type}'")
            if selected_institute_name:
                filters.append(f"institute_name = '{selected_institute_name}'")
            if selected_program_name:
                filters.append(f"program_name = '{selected_program_name}'")
            if selected_category:
                filters.append(f"category = '{selected_category}'")

            # Add filters to the query if any are selected
            if filters:
                query += " AND " + " AND ".join(filters)

            cursor.execute(query)
            result_rows = cursor.fetchall()

            # Populate the treeview with data from the query results
            for row in result_rows:
                tree.insert("", tk.END, values=row)

            tree.pack(padx=10, pady=10)  # Place the tree on the window


root = tk.Tk()
app = ProgramExplorerApp(root)
root.mainloop()
```
